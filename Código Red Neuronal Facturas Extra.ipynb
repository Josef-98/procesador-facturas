```python
import yaml
import pathlib
import sys

# 1. Define DATA_DIR como la carpeta que contiene images/, labels.txt y donde se escribirÃ¡ invoice_data.yaml
DATA_DIR = pathlib.Path(
    "/content/drive/MyDrive/TFM UCM/3. CÃ³digo/3.2 Redes/"
    "facturas/5. Facturas unificado extra"
)

# 2. Comprueba que labels.txt exista en DATA_DIR
labels_path = DATA_DIR / "labels_reducido.txt"
if not labels_path.exists():
    sys.exit(f"âŒ No he encontrado labels.txt en:\n  {labels_path}")

# 3. Lee todas las etiquetas de labels.txt
with labels_path.open("r", encoding="utf-8") as f:
    labels = [line.strip() for line in f if line.strip()]

# 4. Construye el dict de nombres: {0: "Etiqueta0", 1: "Etiqueta1", â€¦}
names_dict = {i: label for i, label in enumerate(labels)}

# 5. Monta el diccionario YAML
yaml_dict = {
    "path": str(DATA_DIR),
    "train": "images",
    "val": "images",
    "names": names_dict
}

# 6. Escribe invoice_data.yaml en la misma carpeta
yaml_path = DATA_DIR / "invoice_data.yaml"
with yaml_path.open("w", encoding="utf-8") as f:
    yaml.dump(yaml_dict, f, sort_keys=False, allow_unicode=True)

print("âœ… YAML vÃ¡lido escrito en:", yaml_path)
print("--- contenido de invoice_data.yaml ---")
print(yaml_path.read_text(encoding="utf-8"))
```

```python
import glob, os

label_dir = "/ruta/a/labels"
for path in glob.glob(os.path.join(label_dir, "*.txt")):
    lines = []
    with open(path) as f:
        for l in f:
            parts = l.split()
            parts[0] = str(int(parts[0]) - 1)
            lines.append(" ".join(parts))
    with open(path, "w") as f:
        f.write("\n".join(lines) + "\n")
```
